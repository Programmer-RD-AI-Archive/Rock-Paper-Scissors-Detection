```python
train_datagen = ImageDataGenerator(rescale=1/255.0,rotation_range=0.2,width_shift_range=0.2,height_shift_range=0.2,zoom_range=0.2,horizontal_flip=True,vertical_flip=True)
test_datagen = ImageDataGenerator(rescale=1/255.0)
```

```python
train_dir = './Rock-Paper-Scissors/train/'
test_dir = './Rock-Paper-Scissors/test/'
```

```python
train_data = train_datagen.flow_from_directory(train_dir,target_size=(224,224))
test_data = train_datagen.flow_from_directory(test_dir,target_size=(224,224))
```

```python
len(train_data)
```

```python
len(test_data)
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
wandb.init(project='RPC',name='Big-Dog-Model')
model_0 = Sequential([
  Conv2D(10,3,padding='same',input_shape=(224,224,3),activation=relu),
  MaxPool2D(),
  Conv2D(10,3,padding='same',activation=relu),
  MaxPool2D(),
  Conv2D(10,3,activation=relu,padding='same'),
  Flatten(),
  Dense(3,activation=softmax)
])
model_0.compile(loss=CategoricalCrossentropy(),metrics=['accuracy'],optimizer=Adam())
model_0.fit(train_data,validation_data=(test_data),epochs=round(12.5),callbacks=[WandbCallback()])
model_0.evaluate(test_data)
model_0.save('/content/drive/MyDrive/Colab Notebooks/Projects/TensorFlow/Todo/Rock-Paper-Scissors-Game-CNN/model.h5')
model = tf.keras.models.load_model('/content/drive/MyDrive/Colab Notebooks/Projects/TensorFlow/Todo/Rock-Paper-Scissors-Game-CNN/model.h5')
```

```python
wandb.init(project='RPC',name='Big-Dog-Model')
model_1 = Sequential([
  Conv2D(10,3,padding='same',input_shape=(224,224,3),activation=relu),
  MaxPool2D(),
  Conv2D(10,3,padding='same',activation=relu),
  MaxPool2D(),
  Conv2D(10,3,activation=relu,padding='valid'),
  Conv2D(10,3,padding='same',activation=relu),
  MaxPool2D(),
  Conv2D(10,3,padding='same',activation=relu),
  MaxPool2D(),
  Conv2D(10,3,activation=relu,padding='valid'),
  Flatten(),
  Dense(3,activation=softmax)
])
model_1.compile(loss=CategoricalCrossentropy(),metrics=['accuracy'],optimizer=Adam())
model_1.fit(train_data,validation_data=(test_data),epochs=round(17.5),callbacks=[WandbCallback()])
model_1.evaluate(test_data)
model_1.save('/content/drive/MyDrive/Colab Notebooks/Projects/TensorFlow/Todo/Rock-Paper-Scissors-Game-CNN/model.h5')
model_1 = tf.keras.models.load_model('/content/drive/MyDrive/Colab Notebooks/Projects/TensorFlow/Todo/Rock-Paper-Scissors-Game-CNN/model.h5')
```

```python
wandb.init(project='RPC',name='Big-Dog-Model')
model_1 = Sequential([
  Conv2D(25,3,padding='same',input_shape=(224,224,3),activation=relu),
  MaxPool2D(),
  Conv2D(25,3,padding='same',activation=relu),
  MaxPool2D(),
  Conv2D(25,3,activation=relu,padding='valid'),
  Conv2D(25,3,padding='same',activation=relu),
  MaxPool2D(),
  Conv2D(25,3,padding='same',activation=relu),
  MaxPool2D(),
  Conv2D(25,3,activation=relu,padding='valid'),
  Flatten(),
  Dense(3,activation=softmax)
])
model_1.compile(loss=CategoricalCrossentropy(),metrics=['accuracy'],optimizer=Adam())
model_1.fit(train_data,validation_data=(test_data),epochs=round(15),callbacks=[WandbCallback()])
model_1.evaluate(test_data)
model_1.save('/content/drive/MyDrive/Colab Notebooks/Projects/TensorFlow/Todo/Rock-Paper-Scissors-Game-CNN/model_1.h5')
model_1 = tf.keras.models.load_model('/content/drive/MyDrive/Colab Notebooks/Projects/TensorFlow/Todo/Rock-Paper-Scissors-Game-CNN/model_1.h5')
```
